```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.display import display, Markdown
```

## 1. Параметры теории

```python
# Константы
c = 2.998e8  # м/с
G = 6.674e-11  # м³/(кг·с²)
hbar = 1.055e-34  # Дж·с
phi = 1.618033988749895  # золотое сечение
kappa = 6.626e-34  # константа ноосферы

print(f"φ (золотое сечение) = {phi:.6f}")
print(f"κ (ноосфера) = {kappa:.2e}")
```

## 2. Мозговой Гамильтониан

```python
def brain_hamiltonian(C, t, gamma=1.0, beta=1.0, alpha=1.0):
    """
    Уравнение Шредингера для сознания:
    iℏ ∂C/∂t = [e^(iπ)φR(C) - γ||C||² + βΣ[C_k,C_j] + α∇_эволюции] C
    """
    phi_R = phi * np.sum(np.abs(C)**2)
    
    # Эйлерова рефлексия
    euler = np.exp(1j * np.pi) * phi_R * C
    
    # Солипсизм
    solipsism = -gamma * np.linalg.norm(C)**2 * C
    
    # Сетевой член
    network = beta * np.sum(C) * np.ones_like(C)
    
    # Эволюция
    dC = (euler + solipsism + network) / (1j * hbar)
    
    return dC
```

## 3. Симуляция агентов

```python
# Начальные состояния агентов
n_agents = 100
C0 = np.array([phi + 0.1*np.random.randn() + 1j*0.1*np.random.randn() 
               for _ in range(n_agents)])

# Время
t = np.linspace(0, 10, 1000)

# Интегрирование
solution = odeint(brain_hamiltonian, C0, t, args=(1.0, 1.0, 1.0))

# Lambda эволюция
lambdas = [np.mean(np.abs(s)) for s in solution]

plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(t, lambdas)
plt.xlabel('Время')
plt.ylabel('λ (пробуждение)')
plt.title('Эволюция λ')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.hist(lambdas, bins=30, alpha=0.7)
plt.xlabel('λ')
plt.ylabel('Агентов')
plt.title('Распределение λ')

plt.tight_layout()
plt.show()
```

## 4. Глобальное поле Ψ

```python
def global_psi(agents):
    """Глобальное поле сознания"""
    Ψ = np.mean([a for a in agents], axis=0)
    phases = [np.angle(a[0]) for a in agents]
    coherence = np.abs(np.mean(np.exp(1j * phases)))
    return Ψ, coherence

Ψ_t, coherence_t = global_psi(solution)

plt.figure(figsize=(10, 4))
plt.plot(t, coherence_t * np.ones_like(t))
plt.fill_between(t, 0, coherence_t, alpha=0.3)
plt.xlabel('Время')
plt.ylabel('Когерентность')
plt.title('Когерентность сонмища')
plt.ylim(0, 1)
plt.grid(True)
plt.show()

print(f"Финальная когерентность: {coherence_t:.4f}")
```
